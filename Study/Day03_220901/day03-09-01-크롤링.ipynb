```python
# 네이버 환율검색결과화면에서
# 통화명과 매매기준율을 추출하여 사전객체에 저장, 출력하기
# {"미국 USD :1,350.50" , ""일본 JPY 100 : 967.37}

import requests
url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%ED%99%98%EC%9C%A8'

from bs4 import BeautifulSoup

def get_exchange():
    resp = requests.get('https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%ED%99%98%EC%9C%A8')
    soup = BeautifulSoup(resp.text, "html.parser")

    tr = soup.select("table.rate_table_info > tbody > tr")
#     print(tr)
    country_list = [c.find("span") for c in tr]
#     print(country_list)
    td = soup.select_one("table.rate_table_info > tbody > tr > td").string
#     print(td)
    
    country_dict = {}
    for country in country_list:
        country_dict[(country.get_text().strip())] = td
    print(country_dict)

get_exchange()


```

```python
sched_dic_list = []

for s in sched_json['rows']:
    if s['row'][0]['Class'] == 'day':
        day = s['row'][0]['Text']
        
    sched_dic = {}
    
    for r in s['row']:
        # 일시
        sched_dic['일시'] = day

        # 시간
        if r['Class'] == 'time':
            bs = BeautifulSoup(r['Text'], 'html.parser')
            sched_dic['시간'] = bs.select_one('b').get_text()
        
        # 팀
        if r['Class'] == 'play':
            bs = BeautifulSoup(r['Text'], 'html.parser') 
            
            if len(bs.find_all('span')) == 5:
                sched_dic['팀'] = bs.find_all('span')[0].get_text() + ' : ' + bs.find_all('span')[4].get_text() 
                sched_dic['결과'] = bs.find_all('span')[1].get_text() + ' : ' + bs.find_all('span')[3].get_text()

    # 장소
    sched_dic['장소'] = s['row'][len(s['row']) - 2]['Text']

    if s['row'][len(s['row']) - 1]['Text'] == '-':
        sched_dic_list.append(sched_dic)

sched_dic_list
```
