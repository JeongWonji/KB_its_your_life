```python
cond = ti['who'] == 'man'
cond
# boolean index를 만들어 조건에 맞는 데이터만 추출 가능
```

```python
ti[cond].head()
```

```python
ti[cond]['age'] = 10
ti
#값 대입할떄 할당이 안돼~! loc 사용해라~
```

```python
# .loc를 쓰지 않으면 값을 추출만 가능하고 값 할당은 불가능하다

ti.loc[cond, 'age'] = 10
# condition(who가 man인 사람들)의 age를 모두 10으로 바꾸겠다

ti
```

```python
cond1 = (ti['age'] >= 20) & (ti['age'] < 40)
cond2 = (ti['pclass'] == 1) | (ti['pclass'] == 2)
cond2
ti.loc[cond1 & cond2, ['survived', 'pclass','age','fare']].head(10)
```
