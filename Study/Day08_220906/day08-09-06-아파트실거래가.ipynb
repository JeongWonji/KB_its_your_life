```python
# 시군구에서 구를 분리, 구칼럼에 저장
apt_price["구"] = apt_price["시군구"].str.split().str[1] 
# 시군구에서 동을 분리, 동 칼럼에 저장
apt_price["동"] = apt_price["시군구"].str.split().str[2] 
apt_price["동"]
```

```python
#거래건수가 제일많은 구 이름과 건수를 출력하세요.
max_name = ""
max_count = 0
# 사용된 구이름 목록 추출(구칼럼.unique())
구이름 = apt_price["구"].unique()
# 구이름을 반복(for)
for 구 in 구이름:
#     print(구,apt_price[apt_price['구'] == 구].shape)
    거래건수 = apt_price[apt_price['구'] == 구].shape[0]
    if 거래건수 > max_count:
        max_count = 거래건수
        max_name = 구
print("거래건수가 가장많은 구",max_name, max_count)
```

```python
계약날짜 = apt_price["계약년월"].astype(str) + apt_price["계약일"].astype(str)
# pd.to_datetime(계약날짜, format = "%Y%m%d")
apt_price["계약요일"] = pd.to_datetime(
    계약날짜, format="%Y%m%d").dt.dayofweek
```

```python
#평방미터를 평으로 변환
apt_price["평"] = apt_price["전용면적(㎡)"] / 3.03
```

```python
# 평을 크기구간으로 나누기
# 10평미만, 10평대, 20평대, 30평대, 40평대, 40평이상
pd.cut(  
    apt_price["평"],
    bins=[0, 10, 20, 30, 40, 50, 10000],
    labels="10평미만,10평대,20평대,30평대,40평대,40평이상".split(",")
)
# apt_price["평"]
```

```python
# 평을 크기구간으로 나누기
# 10평미만, 10평대, 20평대, 30평대, 40평대, 40평이상
apt_price["평형"] = pd.cut(  
    apt_price["평"],
    bins=[0, 10, 20, 30, 40, 50, 10000],
    labels="10평미만,10평대,20평대,30평대,40평대,40평이상".split(",")
)
apt_price.head()
```

```python
#거래금액(만원) 의 값을 기준으로
# 1억미만, 5억미만, 10억미만, 20억미만, 50억미만, 50억이상 의 
# 구간으로 나눠보세요. "가격대" 칼럼에 저장
apt_price['가격대'] = pd.cut(
    apt_price['거래금액(만원)'],
    bins=[0,10000,50000,100000,200000,500000,10000000,],
    labels='1억미만,5억미만,10억미만,20억미만,50억미만,50억이상'.split(','),
    right = False
)
apt_price
```

```python
apt_price.groupby('가격대')['거래금액(만원)'].count()
```

```python
# 강남 3구의 평형별 평균 거래가 출력
gangnam_gu = apt_price[apt_price['구'].isin(['강남구', '송파구', '서초구']) ]
gangnam_gu.groupby('평형')['거래금액(만원)'].mean()

# 강남 3구 이외 지역의 평형별 평균거래가 출력
not_gangnam_gu = apt_price[~apt_price['구'].isin(['강남구', '송파구', '서초구']) ]
not_gangnam_gu.groupby('평형')['거래금액(만원)'].mean()
```

```python
# 구별 , 평형별 평균거래금액 출력
apt_price.groupby(["구","평형"])[["거래금액(만원)"]].mean()
```
